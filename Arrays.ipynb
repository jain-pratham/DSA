Bubble Sort is an algorithm that sorts an array from the lowest value to the highest value. = o(n2)

```python
arr = [1,2,3,10,5,4]

arr_len = len(arr)

for i in range(arr_len - 1):
    swapped = False
    for j in range(arr_len - 1 - i):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
            print(arr)
            swapped = True
    if not swapped:
        break

     
```

The Selection Sort algorithm finds the lowest value in an array and moves it to the front of the array. = o(n2)



```python
arr = [23,43,12,32,99,0]

arr_len = len(arr)
for i in range(arr_len - 1):
    min_index = i
    for j in range(i+1, arr_len):
        if arr[j] < arr[min_index]:
            min_index = j
    # min_value = arr.pop(min_index)
    # arr.insert(i,min_value)
    arr[i], arr[min_index] = arr[min_index], arr[i] 
print(arr)
```

The Insertion Sort algorithm uses one part of the array to hold the sorted values, and the other part of the array to hold values that are not sorted yet. = o(n2)

```python
arr = [23,43,12,32,99,0]
arr_len = len(arr)

for i in range(1,arr_len):
    insert_index = i
    current_value = arr.pop(i)
    for j in range(i-1,-1,-1):
        if arr[j] > current_value:
            insert_index = j
    arr.insert(insert_index, current_value)
print(arr)
```

The Quicksort algorithm takes an array of values, chooses one of the values as the 'pivot' element, and moves the other values so that lower values are on the left of the pivot element, and higher values are on the right of it. = o(n2)

```python
def quickshort(array, low=0, high=None):
    if high is None:
        high = len(array) - 1

    if low < high:
        pivot_index = partition(array, low, high)
        quickshort(array, low, pivot_index-1)
        quickshort(array,pivot_index+1, high)

def partition(array, low, high):
    pivot = array[high]
    i = low - 1

    for j in range(low, high):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]
    array[i+1], array[high] = array[high], array[i+1]
    return i+1

my_array = [64,34,25,12,12,11,90,5]
quickshort(my_array)
print("Sorted array:", my_array)
```

The Counting Sort algorithm sorts an array by counting the number of times each value occurs. = o(n + k)

```python
def countingSort(arr):
    if not arr:
        return arr
    
    max_val = max(arr)
    count = [0] *( max_val + 1)
    
    for num in arr:
        count[num] += 1
    
    arr[:] = []

    for num, freq in enumerate(count):
        arr.extend([num] * freq)

    return arr

unsortedArr = [4, 2, 2, 6, 3, 3, 1, 6, 5, 2, 3,65]
sortedArr = countingSort(unsortedArr)
print(sortedArr)
```

The Radix Sort algorithm sorts an array by individual digits, starting with the least significant digit (the rightmost one). = o(n2)

```python
myArray = [170, 45, 75, 90, 802, 24, 2, 66]
print("Original array", myArray)
radixArray = [[], [], [], [], [], [], [], [], [], []]
maxVal = max(myArray)
exp = 1

while maxVal // exp > 0:

    while len(myArray) > 0:
        val = myArray.pop()
        radixIndex = (val // exp) % 10
        radixArray[radixIndex].append(val)

    for bucket in radixArray:
        while len(bucket) > 0:
            val = bucket.pop()
            myArray.append(val)

    exp *= 10
print("Sorted array:", myArray)
```

The Merge Sort algorithm is a divide-and-conquer algorithm that sorts an array by first breaking it down into smaller arrays, and then building the array back together the correct way so that it is sorted. = O( n â‹…log n)

```python
def mergeSort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    leftHalf = arr[:mid]
    rightHalf = arr[mid:]

    sortedLeft = mergeSort(leftHalf)
    sortedRight = mergeSort(rightHalf)

    return merge(sortedLeft, sortedRight)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    return result

unsortedArr = [3, 7, 6, -10, 15, 23.5, 55, -13]
sortedArr = mergeSort(unsortedArr)
print("Sorted array:", sortedArr)
```

The Linear Search algorithm searches through an array and returns the index of the value it searches for.

```python
def linearSearch(arr, targetValue):
    for i in range(len(arr)):
        if arr[i] == targetValue:
            return i
    return -1
arr = [3,7,4,9,3]
targetValue = 9

result = linearSearch(arr, targetValue)
if result != -1:
    print("Values",targetValue,"found at index", result)
else:
    print("Value",targetValue,"not found")
```
