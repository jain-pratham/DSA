```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def create_node(data):
    return Node(data)

def print_list(node):
    while node is not None:
        print(node.data, end=" -> ")
        node = node.next
    print("null")

node1 = create_node(3)
node2 = create_node(32)
node3 = create_node(13)
node4 = create_node(2)

node1.next = node2
node2.next = node3
node3.next = node4

print_list(node1)
```
