```python
class Solution:
    def pattern1(self, n):
        for x in range(n):
            for j in range(n):
                print("*", end="")
            print("")

object = Solution()
object.pattern1(5)
```

```python
class Solution:
    def pattern2(self, n):
        for x in range(n,0,-1):
            for j in range(x):
                print("*", end=" ")
            print()

object = Solution()
object.pattern2(5)
```

```python
class Solution:
    def pattern3(self, n):
        for i in range(n):
            for j in range(n - i - 1):
                print(" ", end="")
            for k in range(2 * i + 1):
                print("*", end="")
            print()

object = Solution()
object.pattern3(5)
```

```python
class Solution:
    def countDigit(self, n):
        count = 0
        while n > 0:
            count += 1
            n //= 10
        return count

object = Solution()
print(object.countDigit(100000))
```

```python
class Solution:
    def isPalindrome(self, n):
        rev = 0
        store = n
        while n > 0:
            rev = rev * 10 + n % 10
            n = n // 10
        return rev == store

object = Solution() 
object.isPalindrome(101)
```

```python
class Solution:
    def GCD(self, n1, n2):
        i = 1
        ans = 1
        while i <= n1 and i <= n2:
            if i % n1 and i % n2:
                ans = i
            i += 1 
        return ans
    
object = Solution()
print(object.GCD(12, 15))
```

```python
class Solution:
    def isArmstrong(self, n):
        count = 0
        store = n
        while n > 0:
            count += 1
            n //= 10
        n = store
        sum = 0
        while n > 0:
            sum += (n % 10) ** count
            n //= 10    
        return sum == store
    
object = Solution()
print(object.isArmstrong(153))
```

```python
class Solution:
    def divisors(self, n):
        arr = []
        for x in range(1, n + 1):
            if n % x == 0:
                arr.append(x)
        return arr

object = Solution()
object.divisors(12)
```
